```python
import pandas as pd
import numpy as np
import torch
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module3/data/diamonds.csv",
                 index_col=0)

df.head()
```

```python
y = df['price'].values
x = df['carat'].values
```

```python
y = np.array(y, dtype=np.float32)
x = np.array(x, dtype=np.float32)
```

```python
y_tensor = torch.from_numpy(y)
x_tensor = torch.from_numpy(x)
```

```python
coef = torch.tensor([0.0], dtype=torch.float32, requires_grad=True)
intercept = torch.tensor([0.0], dtype=torch.float32, requires_grad=True)
```

```python
learning_rate = 0.01
```

```python
optimizer = torch.optim.SGD([coef, intercept], lr=learning_rate)
```

```python
for t in range(2000):
   y_pred = coef * x_tensor + intercept
   loss = torch.mean((y_pred - y_tensor) ** 2)
   if t % 100 == 99:
        print(f"Iteration {t+1}: Loss = {loss.item()}")

   loss.backward()
   optimizer.step()
   optimizer.zero_grad()
```
