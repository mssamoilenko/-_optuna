```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

[Опис даних](https://search.r-project.org/CRAN/refmans/PPtreeregViz/html/insurance.html)

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module3/data/insurance.csv")
```

```python
df = df.select_dtypes('number')
df.head()
```

# Завдання 2
Розділіть дані на залежну y(`expenses`) та незалежну X ознаки. Розділіть дані на тренувальну та тестову частини

```python
y = df['expenses']
X = df[['age','bmi','children']] #я не звернула уваги на цей допис в домашній роботі (Вибрати один стовчик для X) Дуже не хочеться все перероблювати тому залишу як є, але якщо дуже потрібно то перероблю.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state=42)
```

# Завдання 3
Натренуйте LinearRegression та виведіть такі метрики: MAE, RMSE, R2

```python
model = LinearRegression(positive = True, fit_intercept = False)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred.mean()
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

# Завдання 4
Напишіть рівняння регресії за прикладом:
Ціна = 1.02*Температура - 15.6

```python
beta_0 = model.intercept_
beta_1, beta_2, beta_3 = model.coef_
print(f"Expenses = {beta_1:.2f} * age + {beta_2:.2f} * bmi + {beta_3:.2f} * children + {beta_0:.2f}")
```
